```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("churn_data_raw.csv")
```

```python
df
```

```python
df.drop(columns=["Notes"], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df.replace('-', np.nan, inplace=True)
```

```python
df
```

```python
shifted_row_index = df[df['Churn'] == '0'].index
```

```python
shifted_indices = [shifted_row_index]
shifted_indices
```

```python
for idx in shifted_indices:
    
    df.loc[idx, 'Payment_Method'] = df.loc[idx, 'Paperless_Billing']
    df.loc[idx, 'Paperless_Billing'] = df.loc[idx, 'Contract']
    df.loc[idx, 'Contract'] = df.loc[idx, 'Tech_Support']
    df.loc[idx, 'Tech_Support'] = df.loc[idx, 'Online_Security']
    df.loc[idx, 'Online_Security'] = df.loc[idx, 'Internet_Service']
    df.loc[idx, 'Internet_Service'] = df.loc[idx, 'Multiple_Lines']
    df.loc[idx, 'Multiple_Lines'] = df.loc[idx, 'Phone_Service']
    df.loc[idx, 'Phone_Service'] = df.loc[idx, 'Customer_Service_Calls']
    df.loc[idx, 'Customer_Service_Calls'] = df.loc[idx, 'Churn']
    df.loc[idx, 'Churn'] = df.loc[idx, 'Monthly_Charges']
    df.loc[idx, 'Monthly_Charges'] = df.loc[idx, 'Total_Charges']
    df.loc[idx, 'Total_Charges'] = np.nan
```

```python
df
```

```python
list_cols_0 = ['Total_Charges', 'Monthly_Charges', 'Customer_Service_Calls']
for col in list_cols_0:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
list_cols_1 = ['Subscription_Start_Date', 'Last_Payment_Date']
for col in list_cols_1:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.loc[df['Age'] > 100, 'Age'] = np.nan
```

```python
df
```

```python
list_cols_2 = ['Age', 'Total_Charges', 'Monthly_Charges']
for col in list_cols_2:
    df[col] = df[col].fillna(df[col].median())
```

```python
df
```

```python
df['Last_Payment_Date']
```

```python
df['Last_Payment_Date'] = df['Last_Payment_Date'].fillna(df['Last_Payment_Date'].max())
```

```python
df['Last_Payment_Date']
```

```python
df.loc[df['Phone_Service'] == 'No', 'Multiple_Lines'] = 'No'
```

```python
df['Gender']
```

```python
df['Gender'] = df['Gender'].fillna('Unknown')
```

```python
df['Gender']
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df
```

```python
df['Customer_Service_Calls'] = df['Customer_Service_Calls'].fillna(0)
```

```python
df
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.title()
```

```python
df
```

```python
df.replace('No Internet Service', 'None', inplace=True)
```

```python
df
```

```python
df['Age'] = df['Age'].astype(int)
df['Customer_Service_Calls'] = df['Customer_Service_Calls'].astype(int)
```
